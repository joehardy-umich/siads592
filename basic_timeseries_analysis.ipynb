```python
import os
import pandas as pd
import altair as alt
import numpy as np
```

```python
weather_data = pd.read_csv('clean_weather_data.csv')
covid_data = pd.read_csv('clean_covid_data.csv')
```

```python
weather_data = weather_data.drop(weather_data.columns[0],axis=1)
covid_data = covid_data.drop(covid_data.columns[0],axis=1)
```

```python
#We can compare average temperature to confirmed cases, deaths, and recoveries
merged_df = covid_data.merge(weather_data,how='inner',left_on=['ObservationDate','Province/State'],right_on=['date','location'])
merged_df
```

```python
covid_ts = merged_df[['date','location','Confirmed']]
tavg_ts = merged_df[['date','location','Average Temperature (deg F)']]
#print(covid_ts)
#covid_ts = covid_ts.set_index(['location','date'])
#covid_ts = covid_ts.sort_index().
covid_ts = covid_ts[covid_ts['Confirmed']!=0.0]
covid_ts = covid_ts.groupby(['location','date']).sum().diff()
covid_ts[covid_ts['Confirmed']<0]=0
# for idx in covid_ts.index.levels[0]:
#     #print(covid_ts['diffs'][idx])
#     #print(covid_ts.loc[idx,'diffs'])
#     covid_ts
#     covid_ts.loc[idx,'diffs'] = covid_ts.loc[idx,'Confirmed'].diff().rename({'Confirmed':'diffs'})
#     print(covid_ts.loc[idx,'Confirmed'].diff().rename({'Confirmed':'diffs'}))
#print(covid_ts)
covid_ts = covid_ts.reset_index()
```

```python
alt.data_transformers.disable_max_rows()
covid_chart = alt.Chart(covid_ts.groupby('date').mean().reset_index()).mark_line(color='black').encode( #
        x=alt.X('date:T',axis=alt.Axis(tickMinStep=180)),y='Confirmed')
tavg_chart = alt.Chart(tavg_ts.groupby('date').mean().reset_index()).mark_line(color='red').encode(
        x=alt.X('date:T',axis=alt.Axis(tickMinStep=180)),y=alt.Y('Average Temperature (deg F)',axis=alt.Axis(titleColor='red')))

(covid_chart+tavg_chart).properties(width=500).resolve_scale(y='independent')
```

```python
charts = []
for location in sorted(covid_ts['location'].unique()):
    covid_chart = alt.Chart(covid_ts[covid_ts['location']==location]).mark_line(color='black').encode(
        x=alt.X('date:T',axis=alt.Axis(tickMinStep=180)),y='Confirmed')
    tavg_chart = alt.Chart(tavg_ts[tavg_ts['location']==location]).mark_line(color='red').encode(
        x=alt.X('date:T',axis=alt.Axis(tickMinStep=180)),y=alt.Y('Average Temperature (deg F)',axis=alt.Axis(titleColor='red')))
    charts.append((covid_chart+tavg_chart).properties(width=500,title=location).resolve_scale(y='independent'))
alt.vconcat(*charts)
```

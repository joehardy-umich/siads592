```python
import os
import pandas as pd
import altair as alt
import numpy as np

weather_data = pd.read_csv('clean_weather_data.csv')
covid_data = pd.read_csv('clean_covid_data.csv')

weather_data = weather_data.drop(weather_data.columns[0],axis=1)
covid_data = covid_data.drop(covid_data.columns[0],axis=1)

covid_data

weather_data

#We can compare average temperature to confirmed cases, deaths, and recoveries
merged_df1 = covid_data.merge(weather_data,how='inner',left_on=['ObservationDate','Province/State'],right_on=['date','location'])

merged_df = merged_df1[merged_df1['Confirmed']!=0.0]
merged_df1['date'] = pd.to_datetime(merged_df1['date'])
confirmed = merged_df.groupby(['location','date']).sum().diff()['Confirmed']
confirmed[confirmed<0]=0
confirmed = confirmed.reset_index()
#print(confirmed)
merged_df = merged_df.merge(confirmed,on=['location','date'],how='left').rename(columns={'Confirmed_y':'Confirmed'})
merged_df2 = merged_df1[merged_df1['Deaths']!=0.0]
merged_df['date'] = pd.to_datetime(merged_df['date'])
merged_df2['date'] = pd.to_datetime(merged_df2['date'])
deaths = merged_df2.groupby(['location','date']).sum().diff()['Deaths']
deaths[deaths<0]=0
deaths = deaths.reset_index()
#deaths['date'] = pd.to_datetime
#print(confirmed)
#print(merged_df)

merged_df = merged_df.merge(deaths,on=['location','date'],how='left').rename(columns={'Deaths_y':'Deaths'})
merged_df
```

```python
Country_in_Interest = ['United States'] #United States Only
traffic_data = apple_data_raw[apple_data_raw['region'].isin(merged_df['location'])]
traffic_data = traffic_data.drop(columns=['geo_type','alternative_name','country'])
traffic_data = traffic_data.groupby('region').mean()

traffic_data = traffic_data.reset_index().melt('region')
traffic_data.columns=['State','Date','Traffic']
traffic_data['Date'] = pd.to_datetime(traffic_data['Date'])
#traffic_data = traffic_data[traffic_data['Country'].isin(Country_in_Interest)]
traffic_data = traffic_data[~traffic_data['Date'].isin(['2020-05-11','2020-05-12', '2021-03-12'])]
traffic_data
```

```python
merged_df['date'] = pd.to_datetime(merged_df['date'])
merged_df = merged_df.merge(traffic_data,how='inner',left_on=['date','location'],right_on=['Date','State'])
```

```python
merged_df = merged_df[['date','location','Confirmed','Average Temperature (deg F)','Traffic']]
merged_df = merged_df[merged_df['location'].str.strip()!='']
merged_df
```

```python
charts = []
colors=['black','red','blue']
for i,y_ in enumerate(['Confirmed','Average Temperature (deg F)','Traffic']):
    chart = alt.Chart().mark_line(color=colors[i]).encode(
        y = alt.Y(y_),
        x = 'date')
    charts.append(chart)
charts = [charts[0],alt.layer(charts[1],charts[2]).resolve_scale(y='shared')]
alt.data_transformers.disable_max_rows() 
alt.layer(*charts,data=merged_df).properties(width=200,height=200).resolve_scale(y='independent').facet('location',columns=3).resolve_scale(x='shared',y='independent')
```

```python
alt.data_transformers.disable_max_rows()
#scatter = alt.Chart(loc_timeseries.reset_index()).mark_point().encode(x='date:T',y='mean(Average Temperature (deg F))')
#print(loc_timeseries.reset_index().groupby('date').mean().reset_index())
mean_temp = alt.Chart(merged_df.reset_index().groupby('date').mean().reset_index()).transform_window(rolling_temp='mean(Average Temperature (deg F))',frame=[-7,7]
                                                               ).mark_line(color='red',size=1).encode(
                                                                    x='date:T',
                                                                    y=alt.Y('rolling_temp:Q',axis=alt.Axis(titleColor='red')),
                                                                    )#,scale=alt.Scale(domain=[20,80])))
mean_covid = alt.Chart(merged_df.reset_index().groupby('date').mean().reset_index()).transform_window(rolling_confirmed='mean(Confirmed)',frame=[-7,7]
                                                               ).mark_line(color='black',size=1).encode(
                                                                    x='date:T',
                                                                    y=alt.Y('rolling_confirmed:Q',axis=alt.Axis(titleColor='black')),
                                                                    )

mean_traffic = alt.Chart(merged_df.reset_index().groupby('date').mean().reset_index()).transform_window(rolling_traffic='mean(Traffic)',frame=[-7,7]
                                                               ).mark_line(color='blue',size=1).encode(
                                                                    x='date:T',
                                                                    y=alt.Y('rolling_traffic:Q',axis=alt.Axis(titleColor='blue')),
                                                                    )
alt.layer((mean_temp+mean_traffic).properties(width=400).resolve_scale(y='shared',x='shared'),mean_covid).resolve_scale(y='independent',x='shared').properties(width=400)
```
